```
#Find three simple bivariate functions f1, f2, and f3 and three points p1, p2, p3 such that

#p1 is a minimum of f1
#p2 is a maximum of f2
#p3 is neither a minimum nor maximum of f3, but the gradient of f3 at p3 is the zero vector.

#Use (a) three-dimensional plots as in three_dimensional_plotting.ipynb and 
#(b) density plots and (c) contour plots as in 
#density_and_contour_plots.ipynb. Indicate clearly the points p1, p2, and p3 
#in all plots.

%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

import numpy as np
from scipy.stats import gaussian_kde
```

$f_1 (x,y) = x^3 - 12xy + 8y^3$

local minimum = $-8$ at point ($2,1$)

```
def f1(x, y):
  return ((x**3) - (12*x*y) + (8*(y**3)))

#f1(x,y) = ((x**3) - (12*x*y) + (8*(y**3)))
#minima = -8 at (2,1)
x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(x, y)
Z= f1(X, Y)
p1 = np.array([[2],[1]])

fig = plt.figure()
ax = plt.axes(projection= "3d")
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.view_init(7, 10)
plt.plot(p1[0],p1[1], 'ro')
plt.show()

plt.contour(X, Y, Z, colors='black');
plt.plot(p1[0],p1[1], 'ro')
plt.show()

data = np.vstack([x,y])
kde = gaussian_kde(data)
plt.imshow(Z.reshape(X.shape),
           origin='lower', aspect='auto',
           extent=[-5, 5, -6, 6],
           cmap='Blues')
plt.plot(p1[0],p1[1], 'ro')
cb = plt.colorbar()
cb.set_label("density")
```

$f_2 (x,y) = xy - x^2 y -x y^2$

local maximum = $\frac{1}{27}$ at point ($\frac{1}{3}$, $\frac{1}{3}$)

```
def f2(x, y):
  return (x*y - ((x**2)*y) - (x*(y**2)))

x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(x, y)
Z = f2(X, Y)
p2 = np.array([[1/3], [1/3]])

fig = plt.figure()
ax = plt.axes(projection= "3d")
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.view_init(40, 10)
plt.plot(p2[0],p2[1], 'bo')
plt.show()

plt.contour(X, Y, Z, colors='black');
plt.plot(p2[0],p2[1], 'bo')
plt.show()

data = np.vstack([x,y])
kde = gaussian_kde(data)
plt.imshow(Z.reshape(X.shape),
           origin='lower', aspect='auto',
           extent=[-5, 5, -6, 6],
           cmap='Greens')
plt.plot(p2[0],p2[1], 'bo')
cb = plt.colorbar()
cb.set_label("density")
```

$f_3 (x,y) = x^2 - y^2$

gradient = $(2x,2y)$ and is the zero vector at point ($0,0$)

```
def f3(x, y):
  return ((x**2) - (y**2))

#f3(x,y) = (x^2 - y^2)
#gradient = (2x, -2y) p3 = (0,0)
x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(x, y)
Z = f2(X, Y)
p3 = np.array([[0], [0]])

fig = plt.figure()
ax = plt.axes(projection= "3d")
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.view_init(40, 10)
plt.plot(p3[0],p3[1], 'go')
plt.show()

plt.contour(X, Y, Z, colors='black');
plt.plot(p3[0],p3[1], 'go')
plt.show()

data = np.vstack([x,y])
kde = gaussian_kde(data)
plt.imshow(Z.reshape(X.shape),
           origin='lower', aspect='auto',
           extent=[-5, 5, -6, 6],
           cmap='Reds')
plt.plot(p3[0],p3[1], 'go')
cb = plt.colorbar()
cb.set_label("density")
```
